```python
#exercise 1 from slides

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,11)
y = x*x

plt.plot(x,y,'bo:')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis(-1,11,-1,101)
plt.grid()
plt.show()

```

```python
#exercise 1 from slides continued...

N = 10
x = np.linspace(0,1.,N)
y = np.cos(x) + np.random.rand(N)

plt.plot(x,y,'bo')
plt.show()

x = np.linspace(0,N,N)
value = np.polyfit(x,y,3)
poly = np.poly1d(value)
plt.plot(x,poly(x),'.')
plt.show()
```

```python
#exercise 2 from slides

fs = 150
f = 5

x = np.arange(fs)
y = np.sin(2*np.pi*f*(x/fs)) 

plt.plot(x,y)
plt.show()

yfft = np.fft.fft(np.sin(2*np.pi*f*(x/fs)))

plt.plot(x,yfft)
plt.show()
```
